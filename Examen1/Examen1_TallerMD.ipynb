## Sección Conceptual

### Pregunta 1: Explica brevemente la diferencia entre overfitting y underfitting y describe un ejemplo de cada uno en el contexto del dataset de atletas.

- **Overfitting:** EL modelo aprende ruidos muy específicos del entrenamniento y nos daría un error muy bajo en entrenamiento y alto en testing. Un ejemplo sereía construir un modelo con muchas combinaciones ente "horas_entrenamiento", "vo2max", "porcentaje_grasa", etc. Funcionaria bien en el training pero no al evaluar.
- **Underfitting:** Es un modelo demasiado simple que no captaría la relación real , nos daríamos cuienta por errores muy grandes en el train y el test. Un ejemplo con este dataset es que si usaramos solo horas_entrenamiento para predecir el "rendimiento" e ignoraramos otras variables indispensables, no se lograría explicar bien el tiempo en 10km con la recta.

### Pregunta 2:  ¿Qué tipo de problemas podría presentar una regresión lineal simple con este dataset? Menciona dos posibles violaciones de sus supuestos y cómo podrías detectarlas.

1. **No linealidad** : La relación entre rendimiento y una sola X puede no ser una recta. Lo mas normal es que tengamos rendimientos irregulares, que disminuyen o se estabilizan a lo largo del camino. Nos daríamos cuenta con una gráfica de dispersión con una recta superpuesta, una curva significaria no linealidad.
2. **Heterocedasticidad** : El error puede ser pequeó en deportistas promedio y aumentar en atletas entrenados, lo detectariamos con un embudo en los residuos si los graficamos contra la predicción.

### Pregunta 3: Explica en tus palabras cómo actúan las penalizaciones Ridge y Lasso, y qué efecto esperarías si las variables horas_entrenamiento y vo2max están muy correlacionadas.

**1.Ridge:** Le pondría un frenado suavce a los coeficientes para que no se vuelvan gigantescos. No apaga ñlas variablñes solo las encoge y repartiria el peso cuando hay variables muy iguales entre si. Si _horas entrenamiento_ y _vo2max_ estuvieran muy correlacionadas, las mantendría con pesos mas pequeños o estables.

**2.Lasso:** Como es mas agresivo, ademas de encoger, podría apagar las variabñes dejandolas en 0. Eso simplificaria el modelo y nos dejaría con lo más importante, eligiria una de las dos y descartaria la otra, lo que nos haria un modelo más compacto pero maybe de una forma arbitraria-

### Pregunta 4: Menciona una situación en la que preferirías usar un modelo no lineal (como un Random Forest o SVR) sobre una regresión lineal en este contexto.

Cuando la relación no sea una recta clara, por ejemplo que el tiempo mejorara rapido al subior el vo2max, tambien cuando hubiera interacciones fuertes entre variables, ejemplo si las horas de entrenamiento cambiaran segun lass lesiones previas o el sexo. También si los errores aumentaran o hubiera outliers. En estos casos usaria RF o SVR que capturarian mejor las curvas.

```python
# Pregunta 6: Ajusta un modelo con Ridge, buscando el hiperparámetro óptimo de regularización con GridSearchCV.

# reutilizamos X, y, X_train, X_test, y_train, y_test definidos antes
num_cols = X.select_dtypes(include=["number"]).columns.tolist()
cat_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()

num_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore"))
])

pre = ColumnTransformer(transformers=[
    ("num", num_pipe, num_cols),
    ("cat", cat_pipe, cat_cols)
])

pipe_ridge = Pipeline(steps=[
    ("pre", pre),
    ("model", Ridge(random_state=42))
])

# buscamos alpha
param_grid = {
    "model__alpha": np.logspace(-3, 3, 13)  
}

cv = KFold(n_splits=5, shuffle=True, random_state=42)

gs = GridSearchCV(
    estimator=pipe_ridge,
    param_grid=param_grid,
    scoring="neg_mean_squared_error",
    cv=cv,
    n_jobs=-1,
    refit=True,
    verbose=0
)

gs.fit(X_train, y_train)

best_alpha = gs.best_params_["model__alpha"]
best_cv_mse = -gs.best_score_
best_cv_rmse = np.sqrt(best_cv_mse)

print(f"Mejor alpha (Ridge): {best_alpha}")
print(f"CV RMSE (5-fold) con mejor alpha: {best_cv_rmse:.3f}")

# evaluacion final con el mejor modelo
pred_train_ridge = gs.predict(X_train)
pred_test_ridge  = gs.predict(X_test)

rmse_train_ridge = np.sqrt(mean_squared_error(y_train, pred_train_ridge))
rmse_test_ridge  = np.sqrt(mean_squared_error(y_test,  pred_test_ridge))

print(f"RMSE Ridge (train): {rmse_train_ridge:.3f}")
print(f"RMSE Ridge (test) : {rmse_test_ridge:.3f}")

# comprobacion recalculanmdo base lineal
from sklearn.linear_model import LinearRegression

pipe_lr = Pipeline(steps=[
    ("pre", pre),
    ("model", LinearRegression())
])
pipe_lr.fit(X_train, y_train)
rmse_train_lr = np.sqrt(mean_squared_error(y_train, pipe_lr.predict(X_train)))
rmse_test_lr  = np.sqrt(mean_squared_error(y_test,  pipe_lr.predict(X_test)))

print(f"[Base] RMSE Lineal (train): {rmse_train_lr:.3f}")
print(f"[Base] RMSE Lineal (test) : {rmse_test_lr:.3f}")

mejor = "Ridge" if rmse_test_ridge < rmse_test_lr else "Lineal"
print(f"→ Prefiero: {mejor} (menor RMSE en test)")
```
